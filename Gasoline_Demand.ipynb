```python
import numpy as np
import pandas as pd
import time
from scipy.stats import linregress
from datetime import datetime
import time
from datetime import date, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.lines as mlines
import datetime
```

```python
gasoline_demand = "Weekly_U.S._Product_Supplied_of_Finished_Motor_Gasoline.Aug7.csv"
```

```python
start_date = "2008-01-01"
end_date =  datetime.datetime.now()
# end_date= datetime.datetime.now()
```

```python
#read in csv, remove and rename headers for gasoline storage
gasoline_demand_df =pd.read_csv(gasoline_demand, header=[1,2,3,4])
gasoline_demand_df.columns = gasoline_demand_df.columns.droplevel(0)
gasoline_demand_df.columns = gasoline_demand_df.columns.droplevel(0)
gasoline_demand_df.columns = gasoline_demand_df.columns.droplevel(0)
gasoline_demand_df.rename(columns={'Week of':'Date'}, inplace=True)
gasoline_demand_df.rename(columns={'Weekly U.S. Product Supplied of Finished Motor Gasoline Thousand Barrels per Day':'Gasoline Demand'}, inplace=True)
gasoline_demand_df['Date'] = pd.to_datetime(gasoline_demand_df['Date'], errors='coerce')
gasoline_demand_df = gasoline_demand_df[(gasoline_demand_df['Date']> start_date) & (gasoline_demand_df['Date']< end_date)]
```

```python
# plot crude storage, gasoline storage, and crude prices on one graph with two axes
#having all three lines on one graph can be confusing
fig, ax1=plt.subplots()
ax1.set_xlabel('Date')
ax1.set_ylabel('Gasoline Demand  (mBPD)')
ax1.plot(gasoline_demand_df['Date'],gasoline_demand_df['Gasoline Demand'], color='g')
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax1.tick_params(axis='y')
_=plt.xticks(rotation=45)

blue_line = mlines.Line2D([], [], color='CornflowerBlue', label='Gasoline Demand')
plt.legend(handles=[blue_line], loc='lower left')
```
